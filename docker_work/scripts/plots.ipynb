```python
# import mesa_reader to make its classes accessible
import mesa_reader as mr
import matplotlib.pyplot as plt
```

```python
# make a MesaData object from a history file
h = mr.MesaData('../prems_mla_1.0/LOGS/history.data')

plt.figure()
plt.plot(h.log_Teff, h.log_L)
plt.scatter(h.log_Teff[0], h.log_L[0], color='red', marker='o')

# set axis labels
plt.xlabel('log Effective Temperature')
plt.ylabel('log Luminosity')

# invert the x-axis
plt.gca().invert_xaxis()
# plt.title('H-R Diagram for a post-MS 15$M_{\\odot}$ star')
# plt.show()
plt.savefig('../plots/prems_mla_1.0_HR_diagram.png')
```

### Central temperature vs. central density plot

```python
plt.figure()
plt.plot(h.log_cntr_Rho, h.log_cntr_T)
plt.scatter(h.log_cntr_Rho[0], h.log_cntr_T[0], color='red', marker='o')
plt.xlabel("Log Density")
plt.ylabel("Log Temperature")
# plt.title('Central temperature vs. central density profile for a post-MS 15$M_{\\odot}$ star')
# plt.show()
plt.savefig('../plots/tutorial_t_rho.png')
```

### Central pressure vs. central density plot

```python
# this works even if you only have logRho and logT!
plt.figure()
plt.plot(h.log_cntr_Rho, h.log_cntr_P)
plt.scatter(h.log_cntr_Rho[0], h.log_cntr_P[0], color='red', marker='o')
plt.xlabel("Log Density")
plt.ylabel("Log Pressure")
# plt.title('Central pressure vs. central density profile for a post-MS 15$M_{\\odot}$ star')
# plt.show()
plt.savefig('../plots/tutorial_p_rho.png')
```
